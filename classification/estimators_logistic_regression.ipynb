```python
import pandas as pd
base = pd.read_csv('census.csv')
base.head()
```

```python
base['income'].unique()
```

```python
# func to transform base['income'] in numerical values(0, 1)
def class_converter(label):
    if label == '>50K':
        return 1
    return 0
```

```python
# transforming labels into nums
base['income'] = base['income'].apply(class_converter)
```
