<h2>Goal: Use multiple parameters of the house to predict its value</h2> 

```python
import pandas as pd
base = pd.read_csv('house_prices.csv')
base.head()
```

```python
# see columns names
base.columns
```

```python
util_columns = ['price', 'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']
```

```python
util_columns
```

```python
base = pd.read_csv('house_prices.csv', usecols=util_columns)
```

```python
base.head()
```

<h1>scaling</h1>
<h2>standardzation -> put all values in same scale</h2>
<h2>normalization -> put all values in between 0 and 1</h2>

```python
# normalization method
from sklearn.preprocessing import MinMaxScaler
scaler_x = MinMaxScaler()
base[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']] = scaler_x.fit_transform(base[['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']]) # all x axis parameters
```

```python
base.head()
```

```python
scaler_y = MinMaxScaler()
base[['price']] = scaler_y.fit_transform(base[['price']])
```

```python
base.head()
```

```python
# separating x and y
x = base.drop('price', axis = 1)
y = base.price
```

```python
x.head()
```

```python
y.head()
```

```python
type(y)
```

```python
predictors_columns = util_columns[1:17]
predictors_columns
```

```python
import tensorflow as tf
```

```python
# definition of feature columns
# feature columns are the columns of input for the model
columns = [tf.feature_column.numeric_column(key=c) for c in predictors_columns ]
```

```python
columns[10]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
print(f"x_train: {x_train.shape}, x_test: {x_test.shape}")
```

```python
print(f"y_train: {y_train.shape}, y_test: {y_test.shape}")
```

```python
train_func = tf.estimator.inputs.pandas_input_fn(x = x_train,
                                                y = y_train,
                                                batch_size=32,
                                                num_epochs=None, 
                                                shuffle=True
                                                )
```

```python
test_func = tf.estimator.inputs.pandas_input_fn(x = x_test,
                                                y = y_test,
                                                batch_size=32,
                                                num_epochs=10000, 
                                                shuffle=True
                                                )
```

```python
regressor = tf.estimator.LinearRegressor(feature_columns=columns)
```

<h1>ESTIMATORS METHODS:</h1>

```python
regressor.train(input_fn=train_func, steps=10000)
```

```python
training_metrics = regressor.evaluate(input_fn=train_func, steps=10000)
```

```python
test_metrics = regressor.evaluate(input_fn=test_func, steps=10000)
```

```python
training_metrics
```

```python
test_metrics
```

```python
# values for prediction, taken from test matrix
prediction_func = tf.estimator.inputs.pandas_input_fn(x = x_test, shuffle=False)
```

```python
# predict values from prediction function
predictions = regressor.predict(input_fn=prediction_func)
```

```python
list(predictions)
```

```python
prediction_values = []
for p in regressor.predict(input_fn=prediction_func):
    prediction_values.append(p['predictions'])
```

```python
prediction_values
```

```python
import numpy as np
prediction_values = np.asarray(prediction_values).reshape(-1, 1)
```

```python
type(prediction_values)
```

```python
prediction_values = scaler_y.inverse_transform(prediction_values)
prediction_values
```
